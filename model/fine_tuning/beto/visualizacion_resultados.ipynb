```python
import matplotlib.pyplot as plt
```

```python
epochs = [0]
train_personal_f1 = [0]
train_economic_f1 = [0]
val_personal_f1 = [0]
val_economic_f1 = [0]
```

```python
with open('training_logs_manifesto_completo.txt', 'r') as f:
    for line in f:
        parts = line.strip().split(', ')
        epoch = int(parts[0].split()[1].split('/')[0])
        train_personal = float(parts[1].split(': ')[1])
        train_economic = float(parts[2].split(': ')[1])
        val_personal = float(parts[4].split(': ')[1])
        val_economic = float(parts[5].split(': ')[1])
        
        epochs.append(epoch)
        train_personal_f1.append(train_personal)
        train_economic_f1.append(train_economic)
        val_personal_f1.append(val_personal)
        val_economic_f1.append(val_economic)

```

```python
# Graficar los resultados
fig, axs = plt.subplots(2, 1, figsize=(10, 10))

# Gráfico para F1 Personal
axs[0].plot(epochs, train_personal_f1, label='Train Personal F1', color='blue')
axs[0].plot(epochs, val_personal_f1, label='Validation Personal F1', color='orange')
axs[0].set_ylim([0, 1.05])  # Escalar de 0 a 1
axs[0].set_xlabel('Epochs')
axs[0].set_ylabel('F1 Score')
axs[0].set_title('F1 Score - Personal Liberty')
axs[0].legend()

# Gráfico para F1 Económico
axs[1].plot(epochs, train_economic_f1, label='Train Economic F1', color='green')
axs[1].plot(epochs, val_economic_f1, label='Validation Economic F1', color='red')
axs[1].set_ylim([0, 1.05])  # Escalar de 0 a 1
axs[1].set_xlabel('Epochs')
axs[1].set_ylabel('F1 Score')
axs[1].set_title('F1 Score - Economic Liberty')
axs[1].legend()

# Ajustar diseño
plt.tight_layout()
plt.show()
```
